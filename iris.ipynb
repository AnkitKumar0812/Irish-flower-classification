```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df= pd.read_csv("IRIS.csv")
```

```python
df.head
```

```python
X= df[['sepal_length','sepal_width','petal_length','petal_width']]
X.head
```

```python
y=df['species']
y.head
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_predicted=model.predict(X_test)
```

```python
y_predicted
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predicted)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predicted)
```

```python
dc=pd.DataFrame(X_test)
dc['species'] = y_predicted
dc.head
```

```python
plt.title("Given Dataset")
plt.xlabel("Features")
plt.ylabel("Species")
plt.scatter(df['sepal_length'], df['species'], color = 'red', label = 'Sepal Length')
plt.scatter(df['sepal_width'], df['species'], color = 'blue', label = 'Sepal Width')
plt.scatter(df['petal_length'], df['species'], color = 'green', label = 'Petal Length')
plt.scatter(df['petal_width'], df['species'], color = 'yellow', label = 'Petal Width')
plt.legend()
plt.show()
```

```python
plt.title("Test Dataset")
plt.xlabel("Features")
plt.ylabel("Species")
plt.scatter(dc['sepal_length'], dc['species'], color = 'red', label = 'Sepal Length')
plt.scatter(dc['sepal_width'], dc['species'], color = 'blue', label = 'Sepal Width')
plt.scatter(dc['petal_length'], dc['species'], color = 'green', label = 'Petal Length')
plt.scatter(dc['petal_width'], dc['species'], color = 'yellow', label = 'Petal Width')
plt.legend()
plt.show()
```
